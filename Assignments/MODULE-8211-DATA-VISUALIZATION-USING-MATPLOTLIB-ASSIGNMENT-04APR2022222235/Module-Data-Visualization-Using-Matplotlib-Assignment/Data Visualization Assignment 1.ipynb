```python
# Basic Imports


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Importing Cars.csv file
cars = pd.read_csv (r'P:\DA DS AI\Data Science\Assignments\MODULE-8211-DATA-VISUALIZATION-USING-MATPLOTLIB-ASSIGNMENT-04APR2022222235\Module-Data-Visualization-Using-Matplotlib-Assignment\cars.csv')
cars.head(3)
```

```python
x = cars['model']
y = cars['hp']
```

```python
# Ploting the Graph

fig = plt.figure(figsize=(12, 5))

plt.plot(x, y, linewidth=2.0, linestyle='--', color='b', alpha=0.95,
        marker='*')

plt.title('Model Names vs HorsePower')

plt.xlabel('Models of the cars')

plt.ylabel('Horse-Power of Cars')

plt.show()
```
