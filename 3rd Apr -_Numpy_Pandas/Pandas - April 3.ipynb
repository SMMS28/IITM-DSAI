# Concatenation

```python
# Define two dictionaries containing employee data

data1 = {"Name": ['Jai', 'Prince', 'Gaurav', 'Anuj'],
        "Age": [27, 24, 22, 21],
        "Location": ['Mumbai', 'Kannauj', 'Allahabad', 'Kanpur'],
         "Qualification": ['Msc', 'Bsc', 'Btech', 'Bcom']
        }

data2 = {"Name": ['Abhi', 'Ayush', 'Mike', 'Dhiraj'],
        "Age": [29, 20, 19, 21],
        "Location": ['Chennai', 'Delhi', 'Calcutta', 'Gurgaon'],
         "Qualification": ['Bsc', 'MBBS', 'BA', 'Bcom']
        }
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
concatenated_df = pd.concat([df1, df2])
concatenated_df
```

```python
concatenated_df.reset_index()
```

```python
concatenated_df
```

```python
concatenated_df.reset_index(drop=True)
```

```python
concatenated_df.reset_index(inplace=True, drop=True)
```

```python
concatenated_df
```

# Joins

```python
## Inner Joins

d1 = {'CustomerId': pd.Series([1,2,3,4,5,6]),
     'Product': pd.Series(['Oven', 'Oven', 'Oven',
                          'Television', 'Television', 'Television'])
     }
```

```python
df1 = pd.DataFrame(d1)
```

```python
df1
```

```python

d2 = {'CustomerId': pd.Series([2,4,6,7,8]),
     'State': pd.Series(['California', 'California', 'Texas',
                          'New York', 'Indiana'])
     }
```

```python
df2 = pd.DataFrame(d2)
```

```python
df2
```

```python
df1
```

```python
inner_join_df = pd.merge(df1, df2, on='CustomerId', how='inner')
```

```python
inner_join_df
```

```python
## Outer Join

outer_join_df = pd.merge(df1, df2, on='CustomerId', how='outer')
outer_join_df
```

```python
## Left Join


left_join_df = pd.merge(df1, df2, on='CustomerId', how='left')
left_join_df
```

```python
## Right Join


right_join_df = pd.merge(df1, df2, on='CustomerId', how='right')
right_join_df
```
