1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.

```python
# Basic Imports


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# Importing Cars.csv file
cars = pd.read_csv (r'P:\DA DS AI\Data Science\Assignments\MODULE-8211-DATA-VISUALIZATION-USING-MATPLOTLIB-ASSIGNMENT-04APR2022222235\Module-Data-Visualization-Using-Matplotlib-Assignment\cars.csv')
cars.head(3)
```

```python
x = np.array(cars.cyl)
y = np.array(cars.model)
textprops = {"fontsize":10} # Font size of text in pie chart

plt.pie(x, labels=y, explode = None, radius = 2.9, labeldistance=1.1)

plt.title("Pie chart showing Cyl data of every Car Models", fontsize=30)

plt.legend(loc='lower center', bbox_to_anchor=(4, -0.04), ncol=5, fontsize=12)

plt.show()



```
