```python
# Basic Imports


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Line Plot

```python
x = np.arange(0, 10, 1)
```

```python
x
```

```python
y = 2*x +5 
```

```python
y
```

```python
plt.plot(x, y)
```

```python
# Customization

fig = plt.figure(figsize=(7, 5))

plt.plot(x, y, linewidth=2.0, linestyle='--', color='b', alpha=0.95)

plt.title("Linear Plot Example")

plt.xlabel("X-Axis")

plt.ylabel("Y-Axis")

plt.legend(['Simple Line'], loc='best')

plt.show()
```

```python
# Customization

fig = plt.figure(figsize=(12, 5))

plt.plot(x, y, linewidth=2.0, linestyle='--', color='b', alpha=0.95,
        marker='*')

plt.title("Linear Plot Example")

plt.xlabel("X-Axis")

plt.ylabel("Y-Axis")

plt.legend(['Simple Line'], loc='best')

plt.show()
```

```python
# Add multiple linear plots

x = np.arange(0, 10, 1)

y1 = 2*x + 5

y2 = 4.2*x + 1


fig = plt.figure(figsize=(10, 5))

plt.plot(x, y1, linewidth=1.0, linestyle=':', color='b',
        alpha=0.9, marker='^')

plt.plot(x, y2, linewidth=1.0, linestyle='-', color='y',
        alpha=0.9, marker='o')

plt.title("Linear Plot - Multiple Data")

plt.xlabel("X-Axis")

plt.ylabel("Y-Axis")

plt.legend(['Sales of Suzuki Cars', 'Sales of Toyota Cars'], 
          loc='best')

plt.grid(True)

plt.show()
```

```python
# SUBPLOTS


x = np.arange(0, 10, 1)

y1 = 2*x +5

y2 = 3*x + 10

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Line Plot for Y1')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Line Plot for Y2')

plt.show()



```

# CREATE A PLOT FROM PANDAS DATAFRAME OR EXTERNAL DATA

```python
df = pd.DataFrame.from_dict(
data={
    'Apples': 20, 
    'Mangoes': 25,
    'Lemons': 40,
    'Oranges': 45}, orient='index')


df.reset_index(inplace=True)
df.columns=['Fruits', 'Sales']
df
```

```python
fig = plt.figure(figsize=(5, 5))

plt.plot(df['Sales'], df['Fruits'], linewidth=2.0, 
        linestyle='--', color='g', alpha=0.9,
        marker='*')

plt.legend(['Sales of Fruits for the day'], loc='best')

plt.show()
```

# BAR PLOTS

```python
data = {'apples': 20,
       'mangoes': 40,
       'lemons': 5,
       'oranges': 10}


fruits = list(data.keys())

sales = list(data.values())
```

```python
fruits
```

```python
sales
```

```python
fig = plt.figure(figsize=(7, 4))

plt.bar(fruits, sales)

plt.show()
```

```python
# Customize

plt.figure(figsize=(10, 5))

plt.barh(fruits, sales, color='y', alpha=0.9)

plt.title("Sales of Fruits")

plt.xlabel("Sales")

plt.ylabel("Fruits")

plt.show()
```

```python
# Stacked Bar Plot

x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

# SCATTER PLOT

```python
age = [20, 21, 22, 23, 24, 25, 26]
experience = [5, 3, 2, 9, 1, 8, 4]

fig = plt.figure(figsize=(10, 5))
plt.scatter(age, experience)
plt.show()
```

```python
age = [20, 21, 22, 23, 24, 25, 26]
experience = [5, 3, 2, 9, 1, 8, 4]
salary = [1, 2, 3, 4, 5, 6, 7]

plt.figure(figsize=(10, 7))

plt.scatter(age, experience, c='green', s=100, edgecolors='y', 
           marker='o', alpha=0.8)

plt.scatter(age, salary, c='yellow', s=25, edgecolors='b', 
           marker='*', alpha=0.8)

plt.legend(['Experience', 'Salary'], loc='best')

plt.xlabel("Age")

plt.ylabel("Experience, Salary")

plt.show()
```

```python
age = [20, 21, 22, 23, 24, 25, 26]
experience = [5, 3, 2, 9, 1, 8, 4]
salary = [1, 2, 3, 4, 5, 6, 7]

plt.figure(figsize=(10, 7))

plt.scatter(age, experience, c='green', s=100, edgecolors='y', 
           marker='o', alpha=0.8)

plt.scatter(age, salary, c='yellow', s=25, edgecolors='b', 
           marker='*', alpha=0.8)

plt.legend(['Experience', 'Salary'], loc='best')

plt.xlabel("Age")

plt.ylabel("Experience, Salary")

plt.grid(True)

plt.savefig("Scatter Plot.png")

plt.show()
```

# BOX PLOT

```python
total = [20, 4, 1, 30, 20, 90, 40, 10, 45, 500]

orders = [10, 3, 1, 14, 19, 18, 30, 44, 49, 60]

discount = [1, 3, 4, 5, 9, 1, 8, 7, 49, 78]
```

```python
len(total)
```

```python
len(orders)
```

```python
len(discount)
```

```python
data = list([total, orders, discount])
```

```python
data
```

```python
fig = plt.figure(figsize=(10, 7))

plt.boxplot(data)

plt.title("Box Plot on Sales, Orders and Discounts for Company ABC")

plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))

plt.boxplot(data, showmeans=True)

plt.title("Box Plot on Sales, Orders and Discounts for Company ABC")

plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))

plt.boxplot(data, showmeans=True)

plt.xticks([1,2,3], ['Total', 'Orders', 'Discounts'])

plt.title("Box Plot on Sales, Orders and Discounts for Company ABC")

plt.grid(True)

plt.show()
```

# VIOLIN PLOT

```python
total = [20, 4, 1, 30, 20, 90, 40, 10, 45, 500]

orders = [10, 3, 1, 14, 19, 18, 30, 44, 49, 60]

discount = [1, 3, 4, 5, 9, 1, 8, 7, 49, 78]

data = list([total, orders, discount])
```

```python
data
```

```python
fig = plt.figure(figsize=(10, 7))

plt.violinplot(data)

plt.title("Violin Plot")

plt.xticks([1,2,3], ['Total', 'Orders', 'Discounts'])

plt.grid(True)

plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))

plt.violinplot(data, showmedians=True)

plt.title("Violin Plot")

plt.grid(True)

plt.show()
```

# PIE CHART

```python
pies = ['Electricity', 'Rent', 'Internet', 'EMI']

split = [10, 40, 10, 40]

plt.pie(split, labels=pies, autopct='%1.2f%%', shadow=True)

plt.title("Pie chart showing monthly expenditure")

plt.show()

```

```python
pies = ['Electricity', 'Rent', 'Internet', 'EMI']

split = [10, 40, 10, 40]

separation = [0, 0.1, 0, 0]

plt.pie(split, labels=pies, autopct='%1.2f%%', shadow=True,
       explode=separation)

plt.title("Pie chart showing monthly expenditure")

plt.show()

```

# DONUT CHART

```python
group_names = ['GroupA', 'GroupB', 'GroupC']

group_size = [20, 30, 50]

group_size1 = [5]

a, b, c = [plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]

pie1 = plt.pie(group_size, labels=group_names, autopct='%1.1f%%',
              radius=1.5, explode=(0.05, 0.05, 0.05),
              colors=[a(0.9), b(0.4), c(0.5)])

pie2 = plt.pie(group_size1, radius=0.5, colors='w')

plt.show()
```
